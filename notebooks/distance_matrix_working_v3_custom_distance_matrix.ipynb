```python
def dataframe_to_fastq(df, file_path):
    with open(file_path, 'w') as f:
        for _, row in df.iterrows():
            f.write(f"@{row['id']}\n")         # Header line with ID
            f.write(f"{row['sequence']}\n")    # Sequence line
            f.write("+\n")                     # Plus line
            f.write(f"{row['quality']}\n")     # Quality line
dataframe_to_fastq(df_200, 'cleaned/df_200.fastq')
```

```python
def clean_fastq_file(input_file, output_file):
    """
    Process a FASTQ file and remove special characters (:,-,) and spaces from sequence IDs
    while preserving the FASTQ format.

    Args:
        input_file (str): Path to input FASTQ file
        output_file (str): Path to output FASTQ file
    """
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        # Process file four lines at a time (FASTQ format)
        line_count = 0
        for line in infile:
            # Every 4th line starting with line 1 is a sequence identifier (starts with @)
            if line_count % 4 == 0 and line.startswith('@'):
                # Clean the ID line (remove special characters and spaces)
                cleaned_id = line.strip().replace(":", "").replace("-", "").replace(" ", "")
                outfile.write(cleaned_id + '\n')
            else:
                # Write other lines (sequence, + line, and quality scores) as they are
                outfile.write(line)
            line_count += 1

# Example usage
if __name__ == "__main__":
    input_fastq = "/content/cleaned/df_200.fastq"  # Replace with your input file path
    output_fastq = "/content/cleaned/cleaned_df_200.fastq"  # Replace with desired output file path

    try:
        clean_fastq_file(input_fastq, output_fastq)
        print(f"Successfully processed FASTQ file. Cleaned output saved to: {output_fastq}")
    except FileNotFoundError:
        print("Error: Input file not found!")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

```python
import pandas as pd
import numpy as np

def phylip_to_df(phylip_file):
    # Read the phylip file
    with open(phylip_file) as f:
        # First line contains number of taxa
        n_taxa = len(f.readline().strip())

        # Initialize lists for taxa names and distances
        taxa = []
        distances = np.zeros((n_taxa, n_taxa))

        # Read each line
        for i in range(n_taxa):
            # Split the line into parts
            line = f.readline().strip().split()

            # First element is taxon name
            taxa.append(line[0])

            # Rest are distances - note these are only upper triangle
            # Convert remaining elements to floats
            dist_values = [float(x) for x in line[1:]]

            # Fill in the upper triangle
            distances[i, (i+1):] = dist_values

        # Fill in the lower triangle (matrix is symmetric)
        for i in range(n_taxa):
            for j in range(i+1, n_taxa):
                distances[j,i] = distances[i,j]

    # Convert to DataFrame
    df = pd.DataFrame(distances, index=taxa, columns=taxa)

    return df
```

```python
from Bio import Phylo
import numpy as np
import pandas as pd

def newick_to_distance_matrix(newick_file):
    """
    Convert a Newick tree file to a distance matrix in pandas DataFrame format.

    Parameters:
    newick_file (str): Path to the Newick format tree file

    Returns:
    pandas.DataFrame: Distance matrix with taxa labels as both index and columns
    """
    # Read the tree from the Newick file
    tree = Phylo.read(newick_file, 'newick')

    # Get all terminal nodes (leaves)
    terminals = tree.get_terminals()

    # Create a list of taxa names
    labels = [term.name for term in terminals]
    n_taxa = len(labels)

    # Initialize the distance matrix with zeros
    distance_matrix = np.zeros((n_taxa, n_taxa))

    # Calculate pairwise distances
    for i in range(n_taxa):
        for j in range(i + 1, n_taxa):
            # Calculate the distance between two taxa
            distance = tree.distance(terminals[i], terminals[j])

            # Distance matrix is symmetric
            distance_matrix[i, j] = distance
            distance_matrix[j, i] = distance

    # Convert to pandas DataFrame with labels
    df = pd.DataFrame(distance_matrix, index=labels, columns=labels)

    # # Round the distances to 4 decimal places for cleaner display
    # df = df.round(4)

    return df
```
