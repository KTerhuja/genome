```python
import pandas as pd
from Bio import SeqIO

def fasta_to_dataframe(fasta_file):
    # List to hold sequence ID and sequence pairs
    seq_ids = []
    sequences = []

    # Parse the FASTA file using BioPython
    for record in SeqIO.parse(fasta_file, "fasta"):
        seq_ids.append(record.id)  # Add the sequence ID
        sequences.append(str(record.seq))  # Add the sequence

    # Create a DataFrame
    df = pd.DataFrame({
        'Sequence_ID': seq_ids,
        'Sequence': sequences
    })

    return df

fasta_file_50 = '../unique_species_aligned_fasta/aligned_cleaned_df_50.fasta'
df_50 = fasta_to_dataframe(fasta_file_50)


fasta_file_100 = '../unique_species_aligned_fasta/aligned_cleaned_df_100.fasta'
df_100 = fasta_to_dataframe(fasta_file_100)


fasta_file_200 = '../unique_species_aligned_fasta/aligned_cleaned_df_200.fasta'
df_200 = fasta_to_dataframe(fasta_file_200)
```

```python
df_200_100 = pd.read_csv('../distance_data/distance_matrix_200_100.csv')
df_100_50 = pd.read_csv('../distance_data/distance_matrix_100_50.csv')
```

```python
df_200_100.set_index(df_200_100.columns, inplace=True)
df_100_50.set_index(df_100_50.columns, inplace=True)
```

```python
quadrant = df_200_100.iloc[len(df_200)-1:, :len(df_200)]
```

```python
quadrant_100_50 = df_100_50.iloc[len(df_100)-1:, :len(df_100)]
```

```python
def min_index_gt_zero(col):
    # Filter the column to values greater than zero
    # col_gt_zero = col[col > 0]
    
    # If there are any values greater than zero, return the index of the minimum value
    if not col.empty:
        return col[1:].idxmin()
    else:
        return None  # Return None if no values are greater than zero

# Apply the function to each column
min_indices = quadrant.apply(min_index_gt_zero)
min_indices_100_50 = quadrant_100_50.apply(min_index_gt_zero)



```

```python
link_200_100 = pd.DataFrame(min_indices).reset_index()
link_200_100.columns = ['200 sequences', '100 sequences']
```

```python
link_100_50 = pd.DataFrame(min_indices_100_50).reset_index()
link_100_50.columns = ['100 sequences', '50 sequences']
```
